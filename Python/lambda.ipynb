## Lambda Functions

Lambda functions are anonymous functions(Function without a name) defined using lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions.

```python
#Syntax 
lambda arguments: expression 
```

```python
# Example 1
addition = lambda a,b: a+b
type(addition) # <class 'function'>
addition(2,3) # 5
```

```python
#Example 2
even = lambda num : num % 2 == 0
even(5)
```
