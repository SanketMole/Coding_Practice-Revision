## Array library

In python we use list instead of array, as it more flexible. 
But still in python we have a seperate 'array' library. 


```python
import array 

arr = array.array('i', [1,2,3,4,5])   # Data type of integer i.e 'i' 
print(arr)
arr[0] = 10 
print(arr)
```

```python
# 'i' - integer 
# 'f' - float 
# 'u' - unicode character 
```
