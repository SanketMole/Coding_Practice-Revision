## Bubble Sort algorithm

Bubble sort is the most basic sorting algorithm. Here, each neighbouring element is compared with each other and the greatest is moved toward end of the array. 
(i) Each pass pushes largest element at the right. 
(ii) Algorithm runs for 'n' number of passes, where 'n' is "length of array". 

Time: O(n^2)
Space: O(1)

```python
def bubble_sort(arr):
    swap = False
    for i in range(len(arr)):
        for j in range((len(arr)-i)-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap = True
        if swap == False:
            return arr 
    return arr 

if __name__ == "__main__":
    arr = [4,3,5,2,8,1]
    sorted_array = bubble_sort(arr)
    print("Sorted array: ",sorted_array)

```
