## Insertion Sort Algoritm

Insertion sort is a sorting technique. In this algorithm array is sorted in an increasing order, by incrementing size of sorted array.\
Start from first element.\
Increase size of sorted section of array by one and sort it.\
Repeat until completely sorted.

Time: O(n^2)\
Space: O(1)

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        for j in range(i, 0, -1):
            if arr[j] < arr[j-1]:
                arr[j], arr[j-1] = arr[j-1], arr[j]
            else:
                break
    return arr 

if __name__ == "__main__":
    arr = [32, 23, 44, 65, 11]
    print("Sorted array: ", insertion_sort(arr))
```
