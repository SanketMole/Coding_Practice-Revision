## Selection sort algorithm

Selection sort is a sorting technique in which with every pass the minimum element gets to its sorted position.    
Select the least element and place it at right place.    
Minimize array size and implement the same. 

Time: O(n^2)\
Space: O(1)


```python
def selection_sort(arr):
    for i in range(len(arr)-1):
        min = i
        for j in range(i, len(arr)):
            if arr[j] < arr[min]:
                min = j 
        arr[i], arr[min] = arr[min], arr[i]
    return arr 
if __name__ == "__main__":
    arr = [12,20,5,63,10]
    print("Sorted array: ", selection_sort(arr))
```
