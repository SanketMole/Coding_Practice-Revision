## Map Function

The map() function applies a given function of all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful for transforming data in a list comprehensively. 

The map() function is a powerful tool for applying transformations to iterable data structures. It can be used with regular functions, lambda functions and even multiple iterables, providing a versatile approach to data processing in python. By understanding and utilizing map(), you can write with more efficient and readable code. 

```python
num1 = [1,2,3]
num2 = [4,5,6]

added_numbers = list(map(lambda x,y : x + y , num1, num2))
print(added_numbers)
```

```python
# map() to convert a list of strings to integers 

# Type casting using map() 

str_numbers = ['1', '2', '3']
int_numbers = list(map(int, str_numbers))
print(int_numbers)
```

```python
words = ['aaple', 'cherry', 'banana']
upper_word = list(map(str.upper, words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name':'Krish', 'age': 32},
    {'name':'Jack', 'age':33}
]

names = list(map(get_name,people))
print(names)
```
