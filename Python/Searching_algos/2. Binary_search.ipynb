## Binary Search

Binear search is a better searching technique than Linear search. 
(i) Requires an sorted array. 
(ii) Divides array into subparts and compare the middle element with the target. 

Time : O(log n)  
Space: O(1)

```python
my_list = [12, 34, 44, 50, 68]
target = 44

def binary_search(list, target):
    mid = 0 
    start = 0 
    end = len(list)-1
    while start <= end:
        if list[mid] == target:
            return mid 
        elif list[mid] < target:
            start = mid+1 
            mid =((end + start)//2)
        else:
            end = mid-1
            mid = (end + start) // 2 
        
    return -1 

index = binary_search(my_list, target)
print(index)
     
```
