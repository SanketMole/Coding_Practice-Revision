## Linear Search 

Most basic searching technique is linear search, here we compare each element of an array with the target. 

Time: O(n)  
Space: O(1)

```python
my_list = [10, 23, 45, 70, 11]
target = 70

def linear_search(list, target):
    for i in list: 
        if i == target: 
            return True
    return False 

def linear_search_index(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i 
    return -1
    

result = linear_search(my_list, target)
print("Result : ",result)
index = linear_search_index(my_list, target)
print("Target present at index: ",index)

```
